```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as tree
```

```python
df=pd.read_csv('Breast Cancer.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.diagnosis=le.fit_transform(df[['diagnosis']])
```

```python
df
```

```python
df.isnull().sum()
```

```python
x=df.drop('diagnosis',axis=1)
x
```

```python
y=df[['diagnosis']]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(xtrain, ytrain)
```

```python
pred=clf.predict(xtest)
pred
```

```python
xtest['pred_y']=pred
```

```python
xtest
```

```python
xtest['actual_y']=ytest
```

```python
xtest
```

```python
xtest=xtest.drop(['pred_y','actual_y'],axis=1)
```

```python
xtest
```

```python
pred=clf.predict(xtest)
```

```python
pred
```

***Confusion Matrix***

```python
#Confusion Matrix
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, plot_roc_curve
```

```python
classification_report(ytest,pred)
```

```python
cm=confusion_matrix(ytest,pred)
```

```python
clf.score(xtest,ytest)
```

```python
#True Positive (TP) = 103
#True Negative (TN) = 59
#False Positive (FP) = 4
#False Negative (FN) = 5
```

```python
#Precision or Positive Predicted Value
precision=(103)/(103+4)
precision
```

```python
#True Positive Rate/TPR/Sensitivity/Recall/Hit Rate
TPR=(103)/(103+5)
TPR
```

```python
#False Positive Rate/FPR
FPR=(4)/(4+59)
FPR
```

```python
#F1 Measure
F1=(precision+TPR)/2
F1
```

```python
#Harmonic Mean/F1 Score
F1_Score=(2*precision*TPR)/(precision+TPR)
F1_Score
```

```python
#Specificity/Selectivity/True Negative Rate/ TNR
Specificity=(59)/(59+4)
Specificity
```

```python
#Threat Score/TS/Critical Sucess Index/CSI
CSI=(103)/(103+4+5)
CSI
```

```python
#False Discovery Rate/FDR
FDR=(4)/(103+4)
FDR
```

```python
#Balanced Accuracy
BA=(103+59)/2
BA
```
